```
import pylegend
frame = pylegend.samples.pandas_api.northwind_orders_frame()
```

```
# Single named aggregation
frame.groupby("Ship Name")["Order Id"].aggregate(
    "sum"
).head(5).to_pandas()
```

```
# Multiple aggregations
frame.groupby("Ship Name")["Order Id"].aggregate(
    ["min", "max", "count"]
).head(5).to_pandas()
```
